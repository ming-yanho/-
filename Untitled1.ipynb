```python
import pandas as pd
df = pd.read_csv('C://Users/mingj/Desktop/Telco-Customer-Churn.CSV')
df
```

```python
x = df.loc[:,['customerID','gender','SeniorCitizen','Partner','Dependents','tenure','PhoneService','InternetService','OnlineSecurity','StreamingTV','StreamingMovies','Contract','PaperlessBilling','PaymentMethod','MonthlyCharges','TotalCharges']]
y = df.loc[:,['Churn']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.35)
```

```python
x_train_NoName = x_train[['SeniorCitizen','Partner','Dependents','tenure','PhoneService','PaperlessBilling']]
```

```python
x_train_NoName.head(3)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(max_depth = 3, n_estimators = 10)
rfc.fit(x_train_NoName,y_train)
```

```python
x_test_NoName = x_test[['SeniorCitizen','Partner','Dependents','tenure','PhoneService','PaperlessBilling']]
```

```python
x_test_NoName.head(3)
```

```python
print('The accuracy of Random Forest Classifier on testing set:', rfc.score(x_test_NoName, y_test))
```
